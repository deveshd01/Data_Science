```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[51,54,45,58,46,39,43]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color="green", linewidth=4, linestyle="dashed", marker='D', markersize=15)    
# x and y must have same first dimension  #D=dimond
```

```python
# format string         #Order doesn't matter
plt.plot(x,y, 'g+--', alpha=0.7)                # g+ means green color, '+'as marker, --line      # alpha=trasparancy (0 to 1)  
```

# Bar Chart

```python
import numpy as np
```

```python
company=['Google', 'APPL', 'HCL', 'WBC']
xpos=np.arange(len(company))                      # bcox bar() function don't accept string values
xpos
```

```python
revenue = [56, 41, 32, 47]
profit = [44, 37, 25, 35]
```

```python
plt.bar(xpos,revenue)                       
```

```python
plt.xticks(xpos,company)
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.bar(xpos,revenue, label='revenue')
plt.bar(xpos,profit, label='profit')

plt.legend()
```

```python
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue, width=0.4, label='revenue')
plt.bar(xpos+0.2,profit, width=0.4, label='profit')

plt.legend()
```

```python
plt.yticks(xpos,company)
width = 0.4                     # it is necessary to define width before in horizontal bar chart
plt.barh(xpos-0.2,revenue, width, label='revenue')
plt.barh(xpos+0.2,profit, width, label='profit')

plt.legend()
```

# Histogram

Histograms used when we need to see **frequency in a range**

```python
blood_sugar = [109, 108,  95, 108, 113, 102,  96, 117, 118,  99,114]
plt.hist(blood_sugar, bins=3, rwidth=0.95)       # by default 10 bins(columns)   # rwidth is horizonatal width
```

```python
plt.hist(blood_sugar, bins=[90,100,105,125], rwidth=0.95)   # can define x Axis range by bins
```

```python
men = [109, 108,  95, 108, 113, 102,  96, 117, 118,  99,114]
women = [108, 113, 98,  96, 109, 100,  95,  117, 95,  99,114]

plt.hist([men,women],bins=3,color=['r','g'], label=['men','women'])
plt.xlabel('Range of blood sugar')
plt.ylabel('count of number of persons')
plt.legend()
```

```python
plt.hist([men,women],bins=3,color=['r','g'], label=['men','women'], orientation='horizontal')
plt.show()              # to remove extra text above the chart
```

# Piechart

```python
exp_values = [35,20,15,10,20]
exp_labels = ['home', 'food', 'phone', 'car', 'other']
```

```python
plt.axis='equals'        # to look chart perfect circle
plt.pie(exp_values, labels=exp_labels, autopct='%0.2f%%' , radius=1.5, shadow=True, explode=[0,0,0.3,0,0])   
# autopct='%0.2f%%'  %UPTO 2 Decimals     # explode=[0,0,0.3,0,0] remove a sector outside(start from blue anti-clockwise)

plt.show()                # to remove extra text above the chart
```

```python
plt.pie(exp_values, labels=exp_labels, radius=1.5, autopct='%0.2f%%', startangle=180)      # starting angle of blue
```

```python
plt.pie(exp_values, labels=exp_labels, autopct='%0.2f%%' , radius=1.5,explode=[0,0,0.3,0,0])   
plt.savefig("file_name.png", bbox_inches='tight', pad_inches=1.2, transparent=True)
# bbox_inches so that full image will save    # pad_inches  extra padding around chart  # transparent = No Background
```
