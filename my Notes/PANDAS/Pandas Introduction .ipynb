### merge()

```python
df1=pd.DataFrame({
    'city': ['mumbai','delhi','pune','nagpur'],
    'temp': [34,45,40,30],
})

df2=pd.DataFrame({
    'city': ['pune','mumbai','delhi','bhopal'],
    'humadity': [70,75,80,85]
})
```

```python
pd.merge(df1,df2,on='city')             # bhopal & nagpur  not come in merge
```

```python
pd.merge(df1,df2,on='city', how='outer')  
```

```python
pd.merge(df1,df2,on='city', how='left')      # nagpur present BUT bhopal is not present     # how='right'
```

```python
pd.merge(df1,df2,on='city', how='outer', indicator=True)       # where thid data came from
```

```python
df=pd.DataFrame({
    'city': ['mumbai','delhi','pune'],
    'temp': [34,45,40],
    'humadity': [70,75,80]
})

df2=pd.DataFrame({
    'city': ['mumbai','bhopal', 'nagpur','delhi','pune'],
    'temp': [30,40,50,55,60],
    'humadity': [70,80,90,75,85]
})
```

```python
pd.merge(df,df2,on='city')                # if 2 DF have SAME columns the it automatically add _X  &  _Y on column
```

```python
pd.merge(df,df2,on='city', suffixes=('_left', '_right'))   
```

## DF to SQL Database

**Read write data from Database**<br>
***read_sql_table  ,  read_sql_query*** <br>
***read_sql  ,  to_sql***<br>
**RUN QUERY THROUGH ENGINE**

```python
#  Create an in-memory SQLite database.

from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
```

```python
#  Create a table from scratch with 3 rows.
df = pd.DataFrame({'name' : ['User 1', 'User 2', 'User 3']})  
df
```

```python
df.to_sql('users', con=engine)                              #  Table_name=users
engine.execute("SELECT * FROM users").fetchall()
```

```python
# if we want to append in Database Column names should be same

df1 = pd.DataFrame({'name' : ['User 4', 'User 5']})
df1.to_sql('users', con=engine, if_exists='append')
engine.execute("SELECT * FROM users").fetchall()
```

```python
#  Overwrite the table with just df1.
df1.to_sql('users', con=engine, if_exists='replace',index_label='id')
engine.execute("SELECT * FROM users").fetchall()
```
